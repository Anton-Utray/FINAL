### Censo niveles de educación por alcaldía

```python
edu=pd.read_excel('../FINAL/RAW/CENSOS INEGI/cpv2020_b_cdmx_07_educacion.xlsx', sheet_name=8,skiprows=4, header=1)

```

```python
edu.head()
```

```python
edu.drop(columns=['Entidad federativa'], inplace=True) 
```

```python
edu=edu[edu['Demarcación territorial'] != 'Total'] #esto corresponde al subtotal de acaldias, ie total CDMX
```

```python
edu.head()
```

```python
edu.drop(columns=['Unnamed: 9','Unnamed: 10', 'Unnamed: 11', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 28', 'Unnamed: 29', 'Unnamed: 30'], inplace=True) 
```

```python
edu= edu.drop([0,1,2, 3]).reset_index(drop=True)
```

```python
#Las alcaldías vienen con su codigo territorial antes del nombre de la alcaldía. Usamos regex para quitar los valores numericos y dejar solo texto.

texto = re.compile(r'^\d+\s*(.*)')
edu['Demarcación territorial'] = edu['Demarcación territorial'].str.extract(texto, expand=False)
```

```python
col_unique_counts(edu, 'Demarcación territorial')
```

```python
map_to_alcaldia_id(edu, 'Demarcación territorial')
```

```python
col_unique_counts(edu, 'alcaldia_id')
```

```python
col_names= ['Alcaldia', 'Sexo', 'Grupos de edad', 'Edad desplegada', 'Poblacion', 'Sin Escolaridad', 'Preescolar', 'Primaria', 'Secundaria', 'Estudios técnicos con primaria', 'Estudios técnicos con secundaria', 'Prepa', 'Normal Basica', 'Estudios técnicos con prepa', 'Licenciatura', 'Posgrado', 'No especificado', 'alcaldia_id']
edu.columns=col_names
```

```python
edu=edu[edu['Sexo'] != 'Total'] #quitar filas donde sexo es igual a total (agregado de los 2 sexos)
```

```python
edu= edu[edu['Edad desplegada'] == 'Total'] #quitar las filas de edad desplegada que indiquen edades especificas, dejamos unicamente los totales para luego borrar esta columna y quedarnos con la info de columna 'grupos de edad'
```

```python
col_unique_counts(edu, 'Edad desplegada') #ya nos quedamos con un solo valor aquí, podemos borrar la columna 
```

```python
edu = edu.reset_index(drop=True)
```

```python
edu.drop(columns=['Edad desplegada'], inplace=True) 
```

```python
edu.head() #Ahora solo faltaría quitar los valores de la columna Grupos de edad 'total'
```

```python
edu= edu[edu['Grupos de edad'] != 'Total'] 
```
